# Data Preprocessing

```python
# Calculate date differece from today to 01-jan-2022
today = (dt.date.today() - dt.date(2022,1,1)).days
today

#chnage df
df = df.reset_index()[['Date','Close']].set_index('Date')
```

```python
# create train and test dataset
train_df = df.iloc[:-14]
test_df = df.iloc[-14:]
```

```python
# convert 'Close' data values between 0 and 1
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train,scaled_test = scaler.fit_transform(train_df),scaler.fit_transform(test_df)

#define generator
n_input = 14
n_features = 1
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)

X,y = generator[0]
print(f'Given the Array: \n{X.flatten()}')
print(f'Predict this y: \n {y}')
```

```python
# define model
model = Sequential()
model.add(LSTM(50, activation='relu',input_shape=(n_input, n_features)))
model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.fit(generator,epochs=5)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
last_train_batch = scaled_train[-14:]
last_train_batch = last_train_batch.reshape((1, n_input, n_features))
last_train_batch
```

```python
test_predictions = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))

for i in range(len(test_df)):
    
    # get the prediction value for the first batch
    current_pred = model.predict(current_batch)[0]
    
    # append the prediction into the array
    test_predictions.append(current_pred) 
    
    # use the prediction to update the batch and remove the first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
test_predictions
```

```python
true_predictions = scaler.inverse_transform(test_predictions)

test_df['Predictions'] = true_predictions

# test_df.plot(figsize=(14,5))
```
