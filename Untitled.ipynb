```python
import os
import pandas as pd
import glob
import datetime as dt
import plotly.graph_objects as go
import indicators as ind
```

```python
path = "daily_data"
tocken = 'hbar-usd'
files = glob.glob(f"{path}/{tocken.upper()}/*.csv")
```

```python
import selectors


df = pd.read_csv(files[0])
df = df.rename({"Unnamed: 0":'Date',
            'Open':'Open',
            'High':'High',
            'Low':'Low',
            'Close':'Close',
            'Adj Close':'Adj Close',
            'Volume':'Volume',
            },axis=1)
closing_price = df['Close']
bbupper,bbmiddle, bblower = ind.getBollingerBands(closing_price)

fig = go.Figure()

fig.add_trace(go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close']))

fig.add_trace(go.Line(x=df['Date'],y = bbupper,
              name = 'Bollinger Up'))
fig.add_trace(go.Line(x=df['Date'],y = bblower,
              name='Bollinger Low'))
fig.add_trace(go.Line(x=df['Date'],y = bbmiddle,
              name='Bollinger Middle'))


fig.update_layout(title = f'{tocken}',
                  xaxis_title="Date- Slider",
                  yaxis_title="Price",
                )
fig.show()
#ltc['Close'].plot(figsize=(15,8))
#btc.iloc[:,:-2].plot(figsize=(15,8))
```
